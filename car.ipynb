```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('car_data.csv')
```

```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Selling_type', 'Transmission'])
df
```

```python
current_year = 2024
df['Car_Age'] = current_year - df['Year']
```

```python
df = df.drop(['Car_Name', 'Year'], axis=1)
df
df = df.astype({'Fuel_Type_CNG': int, 'Fuel_Type_Diesel': int, 'Fuel_Type_Petrol': int,
                'Selling_type_Dealer': int, 'Transmission_Manual': int})
df
```

```python
x = df.drop('Selling_Price', axis=1).values
y = df['Selling_Price'].values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=.2)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1500)
model.fit(xtrain,ytrain)
res = model.predict(xtest)
```

```python
res = res.reshape(len(res),1)
ytest = ytest.reshape(len(ytest),1)
```

```python
print(np.concatenate((res,ytest),axis=1))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(ytest, res)
print(f'R-squared score: {r2}')

mse = mean_squared_error(ytest, res)
print(f'Mean Squared Error: {mse}')







```
